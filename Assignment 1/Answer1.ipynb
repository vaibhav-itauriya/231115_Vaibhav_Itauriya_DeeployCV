Importing required libraries

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Question Code

```python
def show(name, n, m, i, title, img):
    plt.subplot(n, m, i)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
```

Display the image

```python
img = mpimg.imread('image.jpg')

show("Image 1", 1, 2, 1, "Hills", img)
plt.show()
```
