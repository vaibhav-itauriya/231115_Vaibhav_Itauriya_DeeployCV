# YOLOv5 Model

Importing required Libraries

```python
import cv2
import torch
from ultralytics import YOLO
```

Load YOLOv5 model from PyTorch Hub

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s') # or yolov5m, yolov5l, yolov5x, custom
# model = YOLO('yolov8n.pt')  # Load the model
model.to('cpu')
```

Open video source: 0 for webcam

```python
video_path = 0
cap = cv2.VideoCapture(video_path)
```

Main Code Execution

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame (OpenCV uses BGR, YOLOv5 uses RGB)
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Perform inference
    results = model(frame_rgb)

    # Render results on the frame
    annotated_frame = results.render()[0]

    # Convert back to BGR for OpenCV display
    annotated_frame = cv2.cvtColor(annotated_frame, cv2.COLOR_RGB2BGR)

    # Display the frame
    cv2.imshow('YOLOv5 Object Detection', annotated_frame)

    # Break on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# SSD300_VGG16 Model

Importing Required Libraries

```python
import torch
import torchvision.transforms as T
import cv2
import numpy as np
from torchvision.models.detection import ssd300_vgg16, SSD300_VGG16_Weights
from torchvision.ops import nms
```

Load the SSD300 model

```python
model = ssd300_vgg16(weights=SSD300_VGG16_Weights.DEFAULT)
model.eval()  # Set to evaluation mode
```

Use CPU

```python
device = torch.device('cpu')
model.to(device)
```

Transformation for input frames

```python
def preprocess_frame(frame):
    """Preprocess frame for SSD300 input."""
    transform = T.Compose([
        T.ToPILImage(),
        T.Resize((300, 300)),  # Resize to SSD300's input size
        T.ToTensor(),
        T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    return transform(frame).unsqueeze(0)
```

Perform object detection

```python
def detect_objects(frame, iou_threshold=0.5, score_threshold=0.2):
    """Detect objects in the frame and filter results."""
    input_tensor = preprocess_frame(frame).to(device)
    with torch.no_grad():
        predictions = model(input_tensor)[0]

    boxes = predictions["boxes"]
    scores = predictions["scores"]
    labels = predictions["labels"]

    # Filter results by score threshold
    keep = scores > score_threshold
    boxes = boxes[keep]
    scores = scores[keep]
    labels = labels[keep]

    # If no valid detections, return empty arrays
    if len(boxes) == 0:
        return np.array([]), np.array([]), np.array([])

    # Apply Non-Maximum Suppression (NMS)
    nms_indices = nms(boxes, scores, iou_threshold)
    boxes = boxes[nms_indices]
    scores = scores[nms_indices]
    labels = labels[nms_indices]

    return boxes.numpy(), labels.numpy(), scores.numpy()
```

COCO class labels
Source: https://debuggercafe.com/ssd300-vgg16-backbone-object-detection-with-pytorch-and-torchvision/

```python
COCO_CLASSES = [
    '__background__', 'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
    'train', 'truck', 'boat', 'traffic light', 'fire hydrant', 'N/A', 'stop sign',
    'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow',
    'elephant', 'bear', 'zebra', 'giraffe', 'N/A', 'backpack', 'umbrella', 'N/A', 'N/A',
    'handbag', 'tie', 'suitcase', 'frisbee', 'skis', 'snowboard', 'sports ball',
    'kite', 'baseball bat', 'baseball glove', 'skateboard', 'surfboard', 'tennis racket',
    'bottle', 'N/A', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 'bowl',
    'banana', 'apple', 'sandwich', 'orange', 'broccoli', 'carrot', 'hot dog', 'pizza',
    'donut', 'cake', 'chair', 'couch', 'potted plant', 'bed', 'N/A', 'dining table',
    'N/A', 'N/A', 'toilet', 'N/A', 'tv', 'laptop', 'mouse', 'remote', 'keyboard', 'cell phone',
    'microwave', 'oven', 'toaster', 'sink', 'refrigerator', 'N/A', 'book',
    'clock', 'vase', 'scissors', 'teddy bear', 'hair drier', 'toothbrush'
]
```

Annotate frame with bounding boxes and labels

```python
def annotate_frame(frame, boxes, labels, scores):
    """Draw boxes, labels, and scores on the frame."""
    for box, label, score in zip(boxes, labels, scores):
        x_min, y_min, x_max, y_max = map(int, box)
        color = (0, 255, 0)  # Green color for bounding boxes
        cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), color, 2)

        # Ensure label index is within bounds
        if 0 <= label < len(COCO_CLASSES):
            label_text = f"{COCO_CLASSES[label]}: {score:.2f}"
        else:
            label_text = f"Unknown: {score:.2f}"

        cv2.putText(frame, label_text, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    return frame
```

Scale bounding boxes back to the original frame size

```python
def rescale_boxes(boxes, original_size, resized_size=(300, 300)):
    """Rescale bounding boxes from resized to original frame dimensions."""
    height, width = original_size
    boxes[:, [0, 2]] *= width / resized_size[0]
    boxes[:, [1, 3]] *= height / resized_size[1]
    return boxes
```

Measure FPS

```python
def calculate_fps(start_time, frame_count):
    """Calculate frames per second."""
    elapsed_time = cv2.getTickCount() - start_time
    fps = frame_count / (elapsed_time / cv2.getTickFrequency())
    return fps
```

Real-time detection, 0 to use webcam

```python
cap = cv2.VideoCapture(0)  # Use webcam
frame_count = 0
start_time = cv2.getTickCount()
```

Main Code Execution

```python
print("Starting real-time object detection. Press 'q' to quit.")
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Failed to capture frame.")
        break

    # Detect objects
    original_size = frame.shape[:2] # Original frame dimensions
    boxes, labels, scores = detect_objects(frame)

    # Rescale bounding boxes if any detections exist
    if len(boxes) > 0:
        boxes = rescale_boxes(boxes, original_size)

    # Annotate frame
    annotated_frame = annotate_frame(frame, boxes, labels, scores)

    # Calculate FPS
    frame_count += 1
    fps = calculate_fps(start_time, frame_count)
    cv2.putText(annotated_frame, f"FPS: {fps:.2f}", (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 255, 0), 1)

    # Display the frame
    cv2.imshow('Object Detection', annotated_frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
